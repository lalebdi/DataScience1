```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_spx_ftse = pd.read_csv("line_chart_data.csv")
df_spx_ftse
```

```python
df_spx_ftse["new_date"] = pd.to_datetime(df_spx_ftse["Date"])
```

```python
df_spx_ftse["new_date"]
```

```python
labels = ["S&P", "FTSE"]
plt.figure(figsize =(20, 8))
plt.plot(df_spx_ftse["new_date"], df_spx_ftse["GSPC500"])
plt.plot(df_spx_ftse["new_date"], df_spx_ftse["FTSE100"])
plt.title("Stocks", fontsize = 14, fontweight = "bold")
plt.xlabel('Returns')
plt.ylabel("Date")
plt.legend(labels = labels, fontsize = "large")
plt.show()
```

```python
df_spx_ftse_h2 = df_spx_ftse[(df_spx_ftse.new_date >= '2008-07-01') & (df_spx_ftse.new_date <= '2008-12-31')]
df_spx_ftse_h2
```

```python
labels = ["S&P", "FTSE"]
plt.figure(figsize =(20, 8))
plt.plot(df_spx_ftse_h2["new_date"], df_spx_ftse_h2["GSPC500"], color = "midnightblue")
plt.plot(df_spx_ftse_h2["new_date"], df_spx_ftse_h2["FTSE100"], color = "crimson")
plt.title("Stocks", fontsize = 14, fontweight = "bold")
plt.xlabel('Returns')
plt.ylabel("Date")
plt.legend(labels = labels, fontsize = "large")
plt.show()
```
