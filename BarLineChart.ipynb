```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import PercentFormatter
sns.set()
```

```python
df_kd = pd.read_csv("bar_line_chart_data.csv")
df_kd
```

```python
fig, ax = plt.subplots(figsize = (10, 7))

ax.bar(df_kd["Year"], df_kd["Participants"], color = "k")
ax.set_ylabel("Number of Participants", weight = "bold")
ax.tick_params(axis = "y", width = 2, labelsize = "large")
ax1 = ax.twinx()
ax1.set_ylim(0, 1)
ax1.yaxis.set_major_formatter(PercentFormatter(xmax = 1.0))
ax1.plot(df_kd["Year"], df_kd["Python Users"], color = "#b60000", marker = "D")
ax1.set_ylabel("Python Users", color = "#b60000", weight = "bold")
ax1.tick_params(axis = "y", colors = "#b60000", width = 2, labelsize = "large")
ax.set_title(" KD Nuggesrts", fontsize = "14", weight = "bold")
```
