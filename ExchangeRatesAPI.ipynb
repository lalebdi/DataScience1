Currency Converter

```python
date = input("Please enter the date or latest: ")
base = input("Convert from: ")
curr = input("Convert to: ")
quan = float(input("How much {} do you want to convert: ".format(base)))

url = base_url + "/" + date + "?base=" + base + "&symbols=" + curr
response = requests.get(url)

if( response.ok is False):
    print("\nErrir {}:".format(response.status_code))
    print(response.json()['error'])
else:
    data = response.json()
    rate = data['rates'][curr]
    
    result = quan*rate
    
    print(result)
```

```python
base_site = "https://itunes.apple.com/search"
```

```python
url = base_site +  "?term=the+beatles&country=us"
requests.get(url)
```

```python
r = requests.get(base_site, params ={"term": "the beatles", "country": "us"})
r.status_code
```

```python
r.url
```

```python
info = r.json()
print(json.dumps(info, indent = 4))
```

```python
info.keys()
```

```python
print(json.dumps(info["results"][0], indent = 4))
```

```python
info["resultCount"]
```

```python
r2 = requests.get(base_site, params ={"term": "the beatles", "country": "us", "limit": 5})
```
