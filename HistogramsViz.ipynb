```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_real_estate = pd.read_csv("histogram_data.csv")
df_real_estate
```

```python
sns.set_style("white")
plt.figure(figsize = (8, 6))
plt.hist(df_real_estate["Price"], bins = 8, color ="#108A99")
plt.title("Dist of prices", fontsize = 14, weight ="bold")
plt.xlabel("Price $")
plt.ylabel("Number of Properties")
sns.despine()
plt.show()
```
