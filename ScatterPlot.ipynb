```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_real_estate = pd.read_csv("scatter_data.csv")
df_real_estate
```

```python
plt.figure(figsize = (12, 8))
scatter = plt.scatter(df_real_estate['Area (ft.)'], df_real_estate["Price"], alpha = 0.6, c = df_real_estate['Building Type'], cmap = "viridis")
plt.legend(*scatter.legend_elements(), loc = "upper left", title = "Building type")
plt.title("relationship of area and price", fontsize = 14, weight = "bold")
plt.xlabel("Area", weight = "bold")
plt.ylabel("Price $")
plt.show()
```

```python
plt.figure(figsize = (12, 8))
sns.scatterplot(df_real_estate["Area (ft.)"], df_real_estate['Price'], hue = df_real_estate['Building Type'], palette =["black", "darkblue", "purple", "pink", "white"], s =100)
plt.title("relationship of area and price", fontsize = 14, weight = "bold")
plt.xlabel("Area", weight = "bold")
plt.ylabel("Price $")
plt.show()
```
