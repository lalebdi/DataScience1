```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sn
```

```python
listings = pd.read_csv("listings.csv", delimiter=',')
listings
```

```python
sn.countplot(x = 'neighbourhood_group', data= listings)
plt.show()
```

```python
sn.barplot(x = 'neighbourhood_group', y = 'price', data = listings)
plt.show()
```

The black bars on each bar denotes the confidance interval.

```python
sn.barplot(x ='neighbourhood_group', y ='price', data = listings, ci = False)
plt.show()
```
