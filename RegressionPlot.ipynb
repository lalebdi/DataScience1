```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
df_ad_exp = pd.read_csv("scatter_plot_ii.csv")
df_ad_exp
```

```python
# plt.figure(figsize = (10, 8))
sns.set(rc ={"figure.figsize": (9, 6)})
sns.regplot(x = "Budget", y = "Sales", data = df_ad_exp, scatter_kws= {'color': 'k'}, line_kws={'color': 'red'})
plt.xlabel("Ad Expen")
plt.ylabel("Sales")
plt.title("Effect of ad Exp on Sales", fontsize = 14, weight = "bold")
plt.show()
```

```python
sns.lmplot(x = "Budget", y = "Sales", data = df_ad_exp, height = 10, scatter_kws= {'color': 'k'}, line_kws={'color': 'red'})
plt.xlabel("Ad Expen")
plt.ylabel("Sales")
plt.title("Effect of ad Exp on Sales", fontsize = 14, weight = "bold")
plt.show()
```
